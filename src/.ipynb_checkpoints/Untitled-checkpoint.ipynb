```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

def plotDF(start,end):
    subDF=DF[start:end]

    ax=subDF.plot(figsize=[15,10])
    ax.set_xticks(np.arange(len(subDF.index)))
    ax.set_xticklabels(subDF.index, rotation=90,fontsize=8)

    plt.title("p3 vs GRCh38 coverage comparison %d-%d" % (start,end),fontsize=18)
    plt.xlabel('id', fontsize=16)
    plt.ylabel('mean_coverage', fontsize=16)

DF=pd.DataFrame.from_csv('/Users/ernesto/projects/IGSR/16_12_16/cov_DF.txt', sep='\t')

a=range(0,len(DF.index),100)
last=(a[len(a)-1],len(DF.index))
lst=zip(a, a[1:])
lst.append(last)

fignum=0
plt.figure(1)      
for i in lst:
    fignum+=1 
    start=i[0]
    end=i[1]
    
    plt.subplot(3,3,fignum)             
    plotDF(start,end)






```
